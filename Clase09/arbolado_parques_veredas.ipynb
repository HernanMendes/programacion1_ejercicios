### Ejercicio 9.8: Comparando especies en parques y en veredas

```python
df_parques = pd.read_csv('../Data/arbolado-en-espacios-verdes.csv')
df_veredas = pd.read_csv('../Data/arbolado-publico-lineal-2017-2018.csv')
```

```python
nombre_especie_parque = 'Tipuana Tipu'
nombre_especie_vereda = 'Tipuana tipu'

df_tipas_parques = df_parques[df_parques['nombre_cie'] == 'Tipuana Tipu']
df_tipas_veredas = df_veredas[df_veredas['nombre_cientifico'] == 'Tipuana tipu']
```

```python
cols_parques = ['altura_tot', 'diametro', 'nombre_cie']
cols_veredas = ['altura_arbol', 'diametro_altura_pecho', 'nombre_cientifico']
df_tipas_parques_seleccionados = df_tipas_parques[cols_parques].copy()
df_tipas_veredas_seleccionados = df_tipas_veredas[cols_veredas].copy()
```

```python
df_tipas_parques_seleccionados.rename(columns = {'altura_tot': 'altura', 'diametro': 'diametro', 'nombre_cie': 'nombre'}, inplace = True)
df_tipas_veredas_seleccionados.rename(columns = {'altura_arbol': 'altura', 'diametro_altura_pecho': 'diametro', 'nombre_cientifico': 'nombre'}, inplace = True)
```

```python
df_tipas_parques_seleccionados['ambiente'] = 'parque'
df_tipas_veredas_seleccionados['ambiente'] = 'vereda'
```

```python
df_tipas = pd.concat([df_tipas_parques_seleccionados, df_tipas_veredas_seleccionados])
```

```python
df_tipas
```

```python
df_tipas.boxplot('diametro',by = 'ambiente')
```

```python
df_tipas.boxplot('altura',by = 'ambiente')
```
